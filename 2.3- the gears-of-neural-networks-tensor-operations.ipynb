**Tensor dot**

**Vector Dot**

```
x = np.random.randint(5, size=(3,))
x
```

```
y = np.random.randint(5, size=(3,))
y
```

```
z = np.dot(x,y)
z
```

**Matrix_vector dot**

```
x = np.random.randint(5, size=(2,3))
x
```

```
y = np.random.randint(5, size=(3))
y
```

```
z = np.dot(x,y)
z
```

```
z.ndim
```

**Transpose**

```
x = np.zeros((300,20))
x = np.transpose(x)
print(x.shape)
```

```
x = np.arange(12)
y = x.reshape(3,4)
y
```

```
z = np.transpose(y)
print(z.shape)
z
```
